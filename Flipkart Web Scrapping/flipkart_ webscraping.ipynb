```python
import requests 
from bs4 import BeautifulSoup
```

```python
url="https://www.flipkart.com/search?q=samsung+mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_0_2&otracker1=AS_QueryStore_HistoryAutoSuggest_0_2&as-pos=0&as-type=HISTORY&as-searchtext=sa"
```

```python
r=requests.get(url)
r.status_code
```

```python
model=soup.findAll("div",{'class':'_4rR01T'})
```

```python
productname=[]
for i in model:
    p=i.text
    productname.append(p)
print(productname)
```

```python
prices=soup.findAll('div',{'class':"_30jeq3 _1_WHN1"})
```
