# Data Visualization

# Plot a graph profit vs quantity

```python
plt.bar(data["Quantity"],data["Profit"])
plt.title("Profit vs Quantity", fontsize=15,color="red")
plt.xlabel("Quantity", fontsize=15)
plt.ylabel("Profit", fontsize=15)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.bar("Sub-Category","Category",data=data,color="green")
plt.show()
```

```python
print(data["State"].value_counts())
plt.figure(figsize=(10,4))
sns.countplot(x=data["State"])
plt.xticks(rotation=90)
plt.show()
```

```python
print(data["Sub-Category"].value_counts())
plt.figure(figsize=(10,4))
sns.countplot(x=data["Sub-Category"])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(data["Quantity"],data["Sales"],color="red")
plt.title("Sales vs Quantity", fontsize=15,color="blue")
plt.xlabel("Quantity", fontsize=15)
plt.ylabel("Sales", fontsize=15)
plt.grid(True)
plt.show()
```

```python
sns.relplot(x="Quantity",y="Profit",hue="Segment",data=data)
plt.show()
```

```python
sns.countplot(x="Segment",data=data,palette="coolwarm")
plt.show()
```

```python
plt.title("Segment",fontsize=20)
data["Segment"].value_counts().plot.pie(autopct="%1.1f%%",shadow=True)
plt.show()
```

```python
sns.countplot(x="Category",data=data,palette="tab10")
plt.show()
```

```python
data.groupby("Segment")[["Profit","Sales"]].sum().plot.bar(color=["red","blue"],figsize=(8,5))
plt.ylabel("Profit/Loss and Sales")
plt.show()
```

```python
plt.title("Region",fontsize=20)
data["Region"].value_counts().plot.pie(autopct="%1.1f%%",shadow=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title("Segment wise sales in each region")
sns.barplot(x="Region",y="Sales",data=data,hue="Segment",palette="coolwarm_r")
plt.xlabel("Region",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title("Segment wise sales in each region")
sns.barplot(x="Region",y="Sales",data=data,palette="coolwarm_r")
plt.xlabel("Region",fontsize=15)
plt.show()
```

```python
data["Region"].value_counts()
```

```python
plt.figure(figsize=(12,8))
plt.title("Segment wise sales in each region")
sns.barplot(x="Region",y="Sales",data=data,hue="Segment",order=data["Region"].value_counts().index,palette="coolwarm_r")
plt.xlabel("Region",fontsize=15)
plt.show()
```

```python
data.groupby("Region")[["Profit","Sales"]].sum().plot.bar(color=["green","orange"],figsize=(8,5))
plt.ylabel("Profit/Loss and Sales")
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.countplot(x="State",data=data,palette='rocket_r',order=data['State'].value_counts().index)
plt.xlabel('States',fontsize=25)
plt.xticks(rotation=70)
plt.show()
```

```python
n_data=["Sales","Quantity","Discount","Profit"]
plt.figure(figsize=(10,5))
sns.heatmap(data[n_data].corr(),annot=True,fmt=".4f",cmap="coolwarm_r",center=0)
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot("Discount","Profit",data=data,color="r",label="Discount")
plt.legend()
plt.show()
```
