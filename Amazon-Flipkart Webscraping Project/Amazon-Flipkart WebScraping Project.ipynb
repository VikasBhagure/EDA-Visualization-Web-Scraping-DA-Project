```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36 Edg/91.0.864.59'}
```

```python
url="https://www.amazon.in/s?k=mobile&i=electronics&crid=7ZZZT2E2AJSG&qid=1643637678&sprefix=mobile%2Celectronics%2C477&ref=sr_pg_1"
page=requests.get(url,headers=headers)
print(page)
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
url_list=[]

for i in range(1,151):
    url_list.append("https://www.amazon.in/s?k=mobile&i=electronics&crid=7ZZZT2E2AJSG&qid=1643637678&sprefix=mobile%2Celectronics%2C477&ref=sr_pg_"+str(i))
url_list
```

```python
Item_name=[]
Item_rating=[]
Item_price=[]
```

```python
Product_name= soup.find_all('span',{'class':"a-size-medium a-color-base a-text-normal"})
Product_name
```

```python
len(Product_name)
```

```python
Product_price=soup.find_all('span',{'class':"a-price-whole"})
Product_price
```

```python
len(Product_price)
```

```python
Product_rating=soup.find_all('i',{'class':"a-icon a-icon-star-small a-star-small-4 aok-align-bottom"})
Product_rating
```

```python
for url in url_list:
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.content,'html.parser')
    
    Product_name= soup.find_all('span',{'class':"a-size-medium a-color-base a-text-normal"})
    Product_rating=soup.find_all('i',{'class':"a-icon a-icon-star-small a-star-small-4 aok-align-bottom"})
    Product_price=soup.find_all('span',{'class':"a-price-whole"})
    for names,ratings,prices in zip (Product_name,Product_rating,Product_price):
        Item_name.append(names.text)
        Item_rating.append(ratings.text)
        Item_price.append(prices.text)
        
    
    
    
    
    
```

```python
df=pd.DataFrame({'Product_Name':Item_name,'Product_Rating':Item_rating,'Product_Price':Item_price},columns=['Product_Name', 'Product_Rating','Product_Price'])
df
```

```python
df.to_csv("D:\DataAnalytic\Python\Project\Project On Class\Amazon-Flipkart Webscraping Project/Amazon_dataset.csv")
```
