```python
url_list=[]

for i in range(1,11):
    url_list.append("https://www.flipkart.com/search?q=+mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i))
url_list
```

```python
Item_name=[]
Item_rating=[]
Item_price=[]
```

```python
Product_name= soup.find_all('div',{'class':"_4rR01T"})
Product_name
```

```python
len(Product_name)
```

```python
Product_price=soup.find_all('div',{'class':"_30jeq3 _1_WHN1"})
Product_price
```

```python
len(Product_price)
```

```python
Product_rating=soup.find_all('div',{'class':"_3LWZlK"})
Product_rating
```

```python
for url in url_list:
    page=requests.get(url,headers=headers)
    soup=BeautifulSoup(page.content,'html.parser')
    
    Product_name= soup.find_all('div',{'class':"_4rR01T"})
    Product_rating=soup.find_all('div',{'class':"_3LWZlK"})
    Product_price=soup.find_all('div',{'class':"_30jeq3 _1_WHN1"})
    for names,ratings,prices in zip (Product_name,Product_rating,Product_price):
        Item_name.append(names.text)
        Item_rating.append(ratings.text)
        Item_price.append(prices.text)
        
    
    
    
    
    
```

```python
df=pd.DataFrame({'Product_Name':Item_name,'Product_Rating':Item_rating,'Product_Price':Item_price},columns=['Product_Name', 'Product_Rating','Product_Price'])
df
```

```python
df.to_csv("D:\\directory\\flipkart_dataset.csv")
```
